```python
def model_confusion_matrix(model, predictors, target, threshold = 0.5):
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    """
    y_pred = model.predict(predictors)
    if len(list(set(y_pred))) != 2:
        y_prob_pred = model.predict(predictors)

        y_pred=[]
        for i in range(0,len(y_prob_pred)):
            if np.array(y_prob_pred)[i] > threshold:
                a=1
            else:
                a=0
            y_pred.append(a)
    else:
        pass

    cm = confusion_matrix(target, y_pred)
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```

```python
import statsmodels.api as sm

LogisticReg1 = sm.Logit(y_train, X_train_with_intercept)
LogisticReg1 = LogisticReg1.fit()

print(LogisticReg1.summary())
```

```python
# Finding Optimal Threshold value
logit_y_pred1 = LogisticReg1.predict(X_train_with_intercept)
fpr, tpr, thresholds = roc_curve(y_train, logit_y_pred1)
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold_logit = round(thresholds[optimal_idx], 3)
optimal_threshold_logit
```

```python
# Fit the model on training and report on testing data
lrmodel = LogisticRegression()
lrmodel.fit(X_train, y_train)
y_predict = lrmodel.predict(X_test)
lrmodel_score = lrmodel.score(X_test, y_test)
print(lrmodel_score)
print(metrics.confusion_matrix(y_test, y_predict))
print(metrics.classification_report(y_test, y_predict))
```

```python
#Observations: We can see the model is ginving an accuracy of 88% which is not bad, but we must run other models aswell
```

```python
#testing data
print('\033[1m{:10s}\033[0m'.format('Confusion Matrix of Logistic Regression on Test Data'))
model_confusion_matrix(LogisticReg1, X_test_with_intercept, y_test, optimal_threshold_logit)
```

```python
# AUC and ROC for the testing data

# predict probabilities
probs = lrmodel.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, probs)
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
# show the plot
plt.title('ROC curve for Logistic Regression on test data')
plt.show()
```
